# Problem Statement

#### We have batting data from the T20 careers [Upto 2019] of 1854 cricketers all across the world. 

#### Goal : Predicting the total runs scored by a batsman by the number of innings he played

## Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

## Import Dataset

```python
df = pd.read_csv('t20_simple.csv')
```

```python
df
```

## Exploratory Data Analysis (EDA) 

```python
plt.xlabel("Inns")
plt.ylabel("Runs")
plt.scatter(df.Inns, df.Runs, color='r', marker='+')
```

## Seperate Input and Output Data

```python
X = df.drop(['Runs'], axis=1)
Y = df.Runs
```

```python
df.shape
```

```python
Y.shape
```

```python
X.shape
```

```python
X
```

```python
Y
```

## Train Test Split

```python
train_X, test_X, train_Y, test_Y = train_test_split(X, Y, test_size=0.2)
```

```python
test_Y
```

## Train a model

```python
kohli = LinearRegression()
kohli.fit(train_X, train_Y)
```

## Create a random prediction

```python
kohli.predict([[37]])
```

## Find the Equauation of the Model

### Equation of a straight regression line 

### y = mx + c

```python
m = kohli.coef_
c = kohli.intercept_
print(m)
print(c)
```

### Let's Plot the Regression Line

```python
x = np.arange(0,100)
y = m*x + c
```

```python
plt.scatter(df.Inns, df.Runs)
plt.plot(x, y, color='r')
```

## Testing the accuracy of the Model on Test Data

```python
kohli.score(test_X, test_Y) #R2 Score
```
